```python
import pandas  as pd
import tushare as ts
df=ts.get_hist_data('600519') #获取贵州茅台全部日k线数据 
# df.to_csv('600519.csv')#导出数据
```

```python
df.describe()
```

```python
from pyecharts import Line, Kline, Overlap
import numpy as np

data= df[['open','close','low','high']]
v1 = data.values
v2 =v1.tolist()
date = data.index
open1 = v1[:,0].tolist()
close = v1[:,1].tolist()
lowest = v1[:,2].tolist()
highest = v1[:,3].tolist()

kline = Kline("K 线图")
kline.add("日K",[time for time in date[-1:0:-1]], v2[-1:0:-1],is_datazoom_show=True)
line_1 = Line()
line_1.add("open", [time for time in date[-1:0:-1]], open1[-1:0:-1])
line_2 = Line()
line_2.add("close", [time for time in date[-1:0:-1]], close[-1:0:-1])
line_3 = Line()
line_3.add("lowest", [time for time in date[-1:0:-1]], lowest[-1:0:-1])
line_4 = Line()
line_4.add("highest", [time for time in date[-1:0:-1]], highest[-1:0:-1])

overlap = Overlap()
overlap.add(kline)
overlap.add(line_1)
overlap.add(line_2)
overlap.add(line_3)
overlap.add(line_4)
overlap
```
